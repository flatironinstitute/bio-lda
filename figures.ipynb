```python
import numpy as np
import matplotlib.pyplot as plt
```

# Load dataset

```python
synthetic = np.load('synthetic.npy')
KC = np.load('KC.npy')
```

```python
fs = 18

iters_synthetic = synthetic.shape[2]
iters_KC = KC.shape[2]
# iters_KC_competing = KC_competing.shape[1]

t_synthetic = np.linspace(1,iters_synthetic,iters_synthetic)
t_KC = np.linspace(1,iters_KC,iters_KC)
# t_KC_competing = np.linspace(1,iters_KC_competing,iters_KC_competing)

fig = plt.figure(figsize=(15,6))

plt.subplot(1, 2, 1)
plt.semilogx(t_synthetic,synthetic[0].mean(axis=0),lw=2,label=r'$\pi_1=0.1$')
plt.fill_between(t_synthetic,np.amin(synthetic[0],axis=0),np.amax(synthetic[0],axis=0),alpha=.1)
plt.semilogx(t_synthetic,synthetic[1].mean(axis=0),lw=2,label=r'$\pi_1=0.2$')
plt.fill_between(t_synthetic,np.amin(synthetic[1],axis=0),np.amax(synthetic[1],axis=0),alpha=.1)
plt.semilogx(t_synthetic,synthetic[2].mean(axis=0),lw=2,label=r'$\pi_1=0.3$')
plt.fill_between(t_synthetic,np.amin(synthetic[2],axis=0),np.amax(synthetic[2],axis=0),alpha=.1)
plt.semilogx(t_synthetic,synthetic[3].mean(axis=0),lw=2,label=r'$\pi_1=0.4$')
plt.fill_between(t_synthetic,np.amin(synthetic[3],axis=0),np.amax(synthetic[3],axis=0),alpha=.1)
plt.semilogx(t_synthetic,synthetic[4].mean(axis=0),lw=2,label=r'$\pi_1=0.5$')
plt.fill_between(t_synthetic,np.amin(synthetic[4],axis=0),np.amax(synthetic[4],axis=0),alpha=.1)
plt.axhline(y=1,c='black',ls='--')
plt.xlim(1e2,iters_synthetic)
plt.xlabel('Iterate',fontsize=fs)
plt.ylabel('Accuracy',fontsize=fs)
plt.title('Synthetic dataset',fontsize=fs)
plt.legend(fontsize=fs)
plt.grid()

plt.subplot(1, 2, 2)
plt.semilogx(t_KC,KC[0].mean(axis=0),lw=2,label=r'$\pi_1=\frac{1}{7}$')
plt.fill_between(t_KC,np.amin(KC[0],axis=0),np.amax(KC[0],axis=0),alpha=.1)
plt.semilogx(t_KC,KC[1].mean(axis=0),lw=2,label=r'$\pi_1=\frac{2}{7}$')
plt.fill_between(t_KC,np.amin(KC[1],axis=0),np.amax(KC[1],axis=0),alpha=.1)
plt.semilogx(t_KC,KC[2].mean(axis=0),lw=2,label=r'$\pi_1=\frac{3}{7}$')
plt.fill_between(t_KC,np.amin(KC[2],axis=0),np.amax(KC[2],axis=0),alpha=.1)
plt.axhline(y=1,c='black',ls='--')
plt.xlim(1e2,iters_KC)
plt.xlabel('Iterate',fontsize=fs)
plt.title('KC dataset',fontsize=fs)
plt.legend(fontsize=fs,loc='lower right')
plt.grid()

# plt.subplot(1, 3, 3)
# plt.semilogx(t_KC_competing,KC_competing.mean(axis=0))
# plt.fill_between(t_KC_competing,np.amin(KC_competing,axis=0),np.amax(KC_competing,axis=0),alpha=.1)
# plt.axhline(y=1,c='black',ls='--')
# plt.xlim(1e2,iters_KC_competing)
# plt.xlabel('Iterate',fontsize=fs)
# plt.ylabel('Accuracy',fontsize=fs)
# plt.title('KC dataset (2 compartments)',fontsize=fs)
# plt.grid()

plt.savefig(f'paper.png',dpi=500,bbox_inches='tight')
```

```python
KC_competing_1_7 = np.load('KC_competing_odors_1_7.npy')
KC_competing_2_6 = np.load('KC_competing_odors_2_6.npy')
KC_competing_3_5 = np.load('KC_competing_odors_3_5.npy')
```

```python
fs = 18

iters_KC_competing = KC_competing_1_7.shape[1]

t_KC_competing = np.linspace(1,iters_KC_competing,iters_KC_competing)

fig = plt.figure(figsize=(8,6))

plt.semilogx(t_KC_competing,KC_competing_1_7.mean(axis=0),label='Odors 1 & 7')
plt.fill_between(t_KC_competing,np.amin(KC_competing_1_7,axis=0),np.amax(KC_competing_1_7,axis=0),alpha=.1)
plt.semilogx(t_KC_competing,KC_competing_2_6.mean(axis=0),label='Odors 2 & 6')
plt.fill_between(t_KC_competing,np.amin(KC_competing_2_6,axis=0),np.amax(KC_competing_2_6,axis=0),alpha=.1)
plt.semilogx(t_KC_competing,KC_competing_3_5.mean(axis=0),label='Odors 3 & 5')
plt.fill_between(t_KC_competing,np.amin(KC_competing_3_5,axis=0),np.amax(KC_competing_3_5,axis=0),alpha=.1)
plt.axhline(y=1,c='black',ls='--')
plt.xlim(1e2,iters_KC_competing)
plt.xlabel('Iterate',fontsize=fs)
plt.ylabel('Accuracy',fontsize=fs)
plt.title('KC dataset (competing MBON compartments)',fontsize=fs)
plt.legend(fontsize=fs)
plt.grid()

plt.savefig(f'competing.png',dpi=500,bbox_inches='tight')
```
